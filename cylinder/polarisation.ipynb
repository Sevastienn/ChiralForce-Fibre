```python
import numpy as np
# import matplotlib.pyplot as plt
# from scipy.constants import c as c0
import fields
```

```python
n_in = 1.45         # refractive index of the cylinder
n_out= 1.00         # refractive index of the surrounding medium
λ0 = 250e-9         # wavelength of the light in vacuum [m]
r0 = 0.32*λ0        # radius of cylinder [m]
basis='pauli'       # basis for the force densities [pauli, em, em_reci, em_curl, pauli_curl]
inside=False        # count the force densities inside the cylinder
k0r0min=1;k0r0max=8 # range of k0r0 values to plot
arrows=8;           # number of arrows in the plot  
pdfs=False          # save pdfs of the force densities
res=250             # resolution of the plot
```

```python
(R,L)=fields.get_amplitudes_from_stokes(0,0,1)
print(R,L)
pdfname='circular.pdf'
(ϵ,E,μ,H,X,Y,Z)=fields.get(r0,λ0,n_in,n_out,R=R,L=L,resolution=res) 
(Wt,ΔW,Wc,Wr,grad_Wt,grad_ΔW,grad_Wc,grad_Wr,ωpt,ωΔp,ωpc,ωpr,Re_Π,Im_Π,ωSt,ωΔS)=fields.observables(λ0,ϵ,E,μ,H,X,Y,Z)
fields.plot_EH(r0,λ0,E,H,X,Y,pdf=pdfs,pdfname=pdfname,arrows=arrows)
```

```python
(nΔW,nWc,nWr,n_2ωΔp,n_2ωpc,n_2ωpr,n_kωSt,n_kωΔS,n_kIm_Π_per_c)=fields.plot_max_forces(λ0,R=R,L=L,k0r0min=k0r0min,k0r0max=k0r0max,N=500,resolution=250/3,samples=50,ℓ=1,zix=0,inside=inside,output=True,unscaled=False,scaled=True)
```

```python
fields.plot_forces_rescaled(r0,λ0,ϵ,μ,Wt,ΔW,Wc,Wr,grad_Wt,grad_ΔW,grad_Wc,grad_Wr,ωpt,ωΔp,ωpc,ωpr,Re_Π,Im_Π,ωSt,ωΔS,X,Y,Z,zix=0,w=.9,nΔW=nΔW,nWc=nWc,nWr=nWr,n_2ωΔp=n_2ωΔp,n_2ωpc=n_2ωpc,n_2ωpr=n_2ωpr,n_kωSt=n_kωSt,n_kωΔS=n_kωΔS,n_kIm_Π_per_c=n_kIm_Π_per_c,arrows=arrows,basis=basis,pdf=pdfs,pdfname=pdfname)
```

```python
import fields
import numpy as np
```

```python
n_in = 1.45         # refractive index of the cylinder
n_out= 1.00         # refractive index of the surrounding medium
λ0 = 250e-9         # wavelength of the light in vacuum [m]
r0 = 0.32*λ0        # radius of cylinder [m]
basis='em_reci'       # basis for the force densities [pauli, em, em_reci, em_curl, pauli_curl]
k0r0min=0;k0r0max=4/np.pi;arrows=8
```

```python
fields.material(n_in=n_in,n_out=n_out)
_=fields.get_neff(rmin=k0r0min,rmax=k0r0max,N=500,neff_fig=True,ℓ=1)
```

```python
(ϵ,E,μ,H,X,Y,Z)=fields.get(r0,λ0,n_in,n_out,R=1,L=1) 
fields.plot_EH(r0,λ0,E,H,X,Y)
(Wt,ΔW,Wc,Wr,grad_Wt,grad_ΔW,grad_Wc,grad_Wr,ωpt,ωΔp,ωpc,ωpr,Re_Π,Im_Π,ωSt,ωΔS)=fields.observables(λ0,ϵ,E,μ,H,X,Y,Z)
fields.plot_forces_rescaled(r0,λ0,ϵ,μ,Wt,ΔW,Wc,Wr,grad_Wt,grad_ΔW,grad_Wc,grad_Wr,ωpt,ωΔp,ωpc,ωpr,Re_Π,Im_Π,ωSt,ωΔS,X,Y,Z,zix=0,w=3,arrows=arrows,basis=basis)
```

```python
(ϵ,E,μ,H,X,Y,Z)=fields.get(r0,λ0,n_in,n_out,R=1,L=0) 
fields.plot_EH(r0,λ0,E,H,X,Y)
(Wt,ΔW,Wc,Wr,grad_Wt,grad_ΔW,grad_Wc,grad_Wr,ωpt,ωΔp,ωpc,ωpr,Re_Π,Im_Π,ωSt,ωΔS)=fields.observables(λ0,ϵ,E,μ,H,X,Y,Z)
fields.plot_forces_rescaled(r0,λ0,ϵ,μ,Wt,ΔW,Wc,Wr,grad_Wt,grad_ΔW,grad_Wc,grad_Wr,ωpt,ωΔp,ωpc,ωpr,Re_Π,Im_Π,ωSt,ωΔS,X,Y,Z,zix=0,w=5,arrows=arrows,basis=basis)
```
